```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import seaborn as sb
import string
from itertools import chain
import numpy as np
```

```python
df = pd.read_csv('merged_eng_only.csv', low_memory=False, lineterminator='\n').sample(1000000, random_state=666)
```

```python
df.head()
```

```python
wordcloud = WordCloud(background_color='black',
                     stopwords = set(STOPWORDS),
                      max_words = 20,
                      max_font_size = 40,
                      random_state=42).generate(str(df['text']))

plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
corpus=' '.join(df.loc[:,'text'])
corpus=[x.lower() for x in corpus.split()]
```

```python
k=20
word_series=pd.DataFrame(data=corpus).value_counts()
keys=[str(*idx) for idx in word_series.index][:k]
values=word_series[:k]
bar = sb.barplot(x=keys, y=values)
bar.set_xticklabels(bar.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

```python
word_series=pd.DataFrame(data=corpus).value_counts()
keys=[str(*idx) for idx in word_series.index if str(*idx) not in set(STOPWORDS)][:k]
values=word_series[:k]
bar = sb.barplot(x=keys, y=values)
bar.set_xticklabels(bar.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

```python
def not_letters(word):
    for c in word:
        if c not in string.ascii_lowercase:
            return True
    return False
texts = list(df['text'].sample(10))
texts = list(chain(t.lower() for t in texts))
texts = ' '.join(texts)
words = texts.split()
filtered = list(filter(not_letters, words))
```
