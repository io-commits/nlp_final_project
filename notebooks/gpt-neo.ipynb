```python
from happytransformer import HappyGeneration, GENTrainArgs

gpt_neo = HappyGeneration("GPT-Neo", "EleutherAI/gpt-neo-125M") 

train_args = GENTrainArgs(num_train_epochs=1, learning_rate=5e-05, batch_size=2) 

gpt_neo.train("../input/nlpfinalproject/sample30m_no_hashtags.csv", args=train_args)
```

```python
from happytransformer import HappyGeneration
# ---------------------------------------------------------
happy_gen = HappyGeneration(model_type="GPT-NEO", model_name="EleutherAI/gpt-neo-125M", load_path="model30m/")
```

```python
from happytransformer import GENSettings

top_k_sampling_settings = GENSettings(do_sample=True, top_k=50, max_length=30, min_length=10)
output_top_k_sampling = happy_gen.generate_text("who will win", args=top_k_sampling_settings)

print (output_top_k_sampling.text)
```

```python
from happytransformer import GENSettings

top_k_sampling_settings = GENSettings(do_sample=False, top_k=50, max_length=30, min_length=10)
output_top_k_sampling = happy_gen.generate_text("who will win", args=top_k_sampling_settings)

print (output_top_k_sampling.text)
```
